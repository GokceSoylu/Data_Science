# Korelasyon Grafikleri

bu zamana kadar tek değişken ile işlem yaptık 

değişken kategorikse (sütun)kutu grafik (barplot), 

sayısal ise histogram, yoğunluk, boxplot ve violin grafikleri kullandık.  

şimdi ize iki değişkenli veri görselleştirme yapıcaz bunun için scatterplot saçılım grafiğini kullanıcaz

```python
import pandas as pd
import seaborn as sns
data=sns.load_dataset("tips")
tip=data.copy()
tip.info()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tip)
```

## boyut ekleme

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tip)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=tip)
#stil değişkeninyle daha belirgin hale getidik
```

sadece görüş kolaylığı için değil boyut eklemke daha fazla değişkenş görmek içinde style kullanılıabilir

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", style="day", data=tip)
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tip)
#daha okunaklı oldu :))
```

sayısal değişken eklerken size keyword ile eklemke mantıkklı. genellikle boyuta uygun olarak okunaklı şekilde görselleştirir

```python
sns.scatterplot(x="total_bill", y="tip", size="size", data=tip)
```

ancak şu haliyşe çokta okunur değil. renk eklmeke için hue keyword kullanalım

```python
sns.scatterplot(x="total_bill", y="tip", size="size", hue="size", data=tip)
```
