```python
import pandas as pd
import seaborn as sns
data=sns.load_dataset("diamonds")
df=data.copy()
df.head()
```

## displot hisplot kdeplot

* distplot ve hisplot sürekli (sayı yani) değişkenlerin histogram grafiğini elde etmek için kullanlır. **histogram/distogram(data_frame.değişken_adi)**

* eğer yoğunluk göstergesi olmasın istiyorsak kde=False deriz 

* sadece yoğunluk olsun istiyorsakta kdeplot(data_frame.değikne_adi) kullanılır. eğrinin alyı taralı olsun istersek fill=True deriz:)


histogram için distplot(data_frame.değişken_adi, kde=bool) fonksiyonu kullanılır Please adapt your code to use either `displot` (a figure-level function with
similar flexibility) or `histplot` (an axes-level function for histograms).

```python
sns.displot(df.price, kde=True) 
```

```python
sns.histplot(df.price, kde=True)
```

```python
sns.kdeplot(df.price)
```

```python
sns.kdeplot(df.price, fill=True)
```

```python
sns.kdeplot(df.price, fill=False)
```
