## İŞ Uygulaması

## Web sitede geçirilen ort süre 170 sn mi?

### Tek Örneklem T Testi

* H0: M=170

* H1: M!=170

```python
import numpy as np
import pandas as pd
import pylab 
import scipy.stats as stats
```

```python
olcumler=np.array([17, 160, 234, 149, 145, 107, 197, 75, 201, 225, 211, 119, 
              157, 145, 127, 244, 163, 114, 145,  65, 112, 185, 202, 146,
              203, 224, 203, 114, 188, 156, 187, 154, 177, 95, 165, 50, 110, 
              216, 138, 151, 166, 135, 155, 84, 251, 173, 131, 207, 121, 120])
```

```python
olcumler[:10]
```

```python
stats.describe(olcumler)
```

## Normallik Varsayımı

garfiklerle ve shapiro testiyle gerçekleştirilebilir

#### Grafik

```python
# Histogram
pd.DataFrame(olcumler).plot.hist();
# evet normal duruyor
```

```python
# qqplot
stats.probplot(olcumler, dist="norm", plot=pylab)
pylab.show()
# sol taraf örnek dağilim
# alt tarf teorik dağilimi ifade eder
```

#### Shapiro-Wilks Testi

hipotezlerimizi belirleyelim

H0 = Örnek dağılım ile teaorik normla dağilim dağilim arasındaki istatistiksel olarak anlamllımbir **farklılık yoktur**

H1=Örnek dağılım ile teaorik normla dağilim dağilim arasındaki istatistiksel olarak anlamllımbir **farklılık vardır**

```python
from scipy.stats import shapiro
shapiro(olcumler)
```

p > 0.5 o zaman h0 red edilemez :))

o zmn dağılım normaldir

### Hipotez testinin uygulanması

```python
stats.ttest_1samp(olcumler,popmean=170)
```

p<0.05 o zaman h0 red!

web sitede geçirilen ortalama zaman 170 saniden farlıdır.

### non parametrik tek örneklem Testi

normalde yukarıdaki veri de sorun yok ancak normallik ten geçene-meseydi. yani shapiro testi negatif olsaydı geçemeseydi

nonparametrik test yapmalıydık. yukarıdaki parametrik ti

```python
from statsmodels.stats.descriptivestats import sign_test
```

```python
sign_test(olcumler,170)
# p value >0.05 o zman hipotez0 red edilemez. zaten biz bunun paremetrik olduğunu biliyoruz
#bu sonuçla ilgilenmiyoruz. sorun yok :))
```
