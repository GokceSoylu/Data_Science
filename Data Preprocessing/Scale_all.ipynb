# DEğişkne Standartlaştırmama / Değişken Dönüşümü

* Veri standartlaştırmada karşılaştırılabilirliğini korururuz
* Dönüşümde ise tamamen dönüştürürüz
* standartlaştımada varyasyon korunur ölçek değişir
* dönüşümde ise ölçek bazen korunur bazen bozulur. çünkü böyle bir kaygımız yok 

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df=pd.DataFrame({"V1":V1,
                "V2":V2,
                "V3":V3})
df=df.astype("float64")
print(df.info())
df
```

```python
# standartlaştıma için scale kullanılır
```

```python
vars=list(df)
```

```python
from sklearn import preprocessing
dff=preprocessing.scale(df) # ort = 0 std = 1 olacak şekilde standartlaştırma yaptık
dff
```

```python
dff.std()
```

## Normalizasyon

0 1 arsına dönüştürülür

```python
preprocessing.normalize(df)
```

## Min Max Dönüşümü
isediğmiz iki değer arasına dönüştürme
burada donuşturucu kullanırız

```python
donusturucu=preprocessing.MinMaxScaler(feature_range=(0,100))
donusturucu.fit_transform(df)
```

Buraya kadar değişkenlerin varyans yapısı korundu ama ölçekleri değişti
daha geniiş daha yayvan oldu ama kendi aralırındaki o yayilim korundu çünkü bir işlem hepsine uygulandı

## Binary Dönüşümü
burada eşik değer belirleyip n-binary ddönüşüm yapıyoruz
yani eşik değerli  0 1 dönüşümü

```python
donusturucu_bin=preprocessing.Binarizer(threshold=5).fit(df)
donusturucu_bin.transform(df)
```

## 0 1 dönüşümü
burada maksat sadece değişkenş numerik hale getrimek

```python
import seaborn as sns
df_titanic=sns.load_dataset("titanic")
df_titanic.head()
```

```python
df_titanic["sex"]=df_titanic["sex"].astype("category") #cat sadece kategorşk ile çalıştığı içi dönüşüm yaptık
df_titanic["new_sex"]=df_titanic["sex"].cat.codes
#bir yüntem daha var
```

```python
label01=preprocessing.LabelEncoder()
df_titanic["newest_sex"]=label01.fit_transform(df_titanic["sex"])
df_titanic
```

## 1 diğerlei 0
belirli değişkemin etkisi daha fdazla olduğunu fark ettiğimizde belirli değiken yada değişkenlerşn 1 diğerlerin sıfır olarak değiştirilmeisdir

```python
tip=sns.load_dataset("tips")
tip.head(20)
```

```python
tip["new_day_label"]=np.where(tip["day"].str.contains("Sun"),1,0)
tip
```

## Çok değişkenli Dönüşüm

```python
label=preprocessing.LabelEncoder()
tip["newst_day"]=label.fit_transform(tip["day"])
tip
#burada sınıfları 1,2,3 diye sınıflandırır ama dah aöncede bahsettiğimiz gibi bu durumda
#sınıfları bir sıray sıraya koymuşuz üstğnlüj belirtmişiz gibi olur 
# # bu yüzden kategorik değişkenş numeriğe dönüştürüken one hot dönüşümünü kullanmak daha mantıklı olur 
```

## One Hot Dönüşümü

```python
pd.get_dummies(tip, columns=["day"], prefix=["day"])
# bu one hot'ı eğer ki iki sınıflı bir değişken için kulllanırsan saçma olur. buna da dummy değişken tuzağı denilir
```

## Sürekli DEğikeni Kategpriğe Çevirme

```python
tip.head()
```

```python
df_tip=tip.select_dtypes(include=["float64","int64"])
label_tip=preprocessing.KBinsDiscretizer(n_bins=[3,2,2], encode="ordinal", strategy="quantile" ).fit(df_tip)
label_tip.transform(df_tip)[0:10]
```

```python
# normald sürekliden kategoriğe dönüştürmek pekte mantıklı bir dönüşüm değildir
```
