# Değişkne Dönüşümleri

```python
import pandas as pd
import numpy as np
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
```

```python
df.info()
```

## 0 1 Dönüşümleri

```python
from sklearn.preprocessing import LabelEncoder
lbe=LabelEncoder()
#yine lbe bir nevi dönüştürü yaptık
```

```python
lbe.fit_transform(df["sex"])
# bu değşikliği veri setinin asıl haline kayıt eymemiz gerekiyor. ancak direkt eski halini değiştirmektense
# yeni değişkne oluşturup ona atamak daha mantıklı. 
```

```python
df["new_sex"]=lbe.fit_transform(df["sex"])
df
```

## Çok sınflı dönüşüm

```python
lbe=LabelEncoder()
```

```python
lbe.fit_transform(df["day"])
#sday ınıflarını yeniden 0 1 2.. şeklinde isimlendirdi. sınıfladı yani
```

```python
#kategorik değişkeni böyle değiştirip asıl data ya atarsan bu yapının bozulmasına yanlılığa sebebiyet verir
#normalde day nominal bir kategorik değişkendi. yani sınıflar arsı bir üstünlük yokstu. nacak bu numerik sınıflandırmada bir üstünlük var gibi olur
#bu durum algoritmaları yanıltır
```

## One hot dönüşümü

```python
df.head()
```

```python
df_one_hot=pd.get_dummies(df, columns=["sex"], prefix=["sex"])
df_one_hot.head()
#bir değişkenş ifade eden başka değişken varsa dummy değişken veri tuzağı olur
```

```python
df_one_hot_day=pd.get_dummies(df,columns=["day"],prefix=["day"])
df_one_hot_day.head()
```

```python
# one hot dönüşümü faydaları
#birincisi değişkenş numerik yapmış oluruz
#değişkenin herhangi bir sınıfındaki ağırlık algorştmaya hissetrilmiş olur

#dummy değişken tuzağı
#eğer one hot dönüşümünü uyguladığın değişken iki sınıflıysa bu anlamsız bir dönüşüm olur 
#iki değişkende birbirinni ayni şeyi ifade edebilir durmda olur buna dummy değiişkne tuzağı deneir. bundan kaçımnılmalıdır.
```
