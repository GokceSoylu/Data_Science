```python
import pandas as pd
import seaborn as sns
data=sns.load_dataset("tips")
tip=data.copy()
tip.info()
```

```python
sns.boxplot(x="total_bill", data=tip);
#sns.boxplot(x=tip["total_bill"]);

#ikiside doğru sorun yok :)
```

## Çaprazlama

```python
tip.describe().T
```

### Hangi günler daha fazla kazanıyoruz

```python
sns.catplot(x="day", y="total_bill", kind="point", data=tip)
```

```python
sns.boxplot(x="day", y="total_bill", data=tip)
#bu grafiğin güzel yani oetalmayıda günlerin aykırı değerleride görünür
```

```python
sns.boxplot(x="time", y="total_bill", data=tip)
#bu iki graafikten çıkarımlarla en iyi elemanlarımı haftasonu ve akşamları çalıştırırm
#gerekirse iyi eleman hafta içi çalışmasın cumartresi çalışsın :)
```

## kişi sayısı kazanç

```python
sns.boxplot(x="size", y="total_bill", data=tip)
```

## gün ve cinsiyetin ödenen hesapla ilişkisi

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tip)
```

## violin

```python
sns.catplot(y="total_bill", kind="violin", data=tip)
#violin grafik biraz daha tercihen kullanılacak grafik
```

```python
sns.catplot(x="day", y="total_bill", data=tip, kind="violin")
#yatay olarak bakıldığınd ayoğunkluk grafiği formu, 
# #dikey olaark bakıldığınd aise box grafik formu var
```
