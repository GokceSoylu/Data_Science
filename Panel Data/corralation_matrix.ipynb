## pairplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("iris")
idt=data.copy()
```

```python
idt.info()
```

```python
idt.species.value_counts()
```

pairplot da direkt data farem in adını yazdığımızda tüm değişkenlerin birbiriyle

ilişkisini gözlemleyebiliriz. 

bunu catplot, scatterplot gibi diiğer fonksiyonlara yazdığımızda tek grafikte göstrilir 

değişkenlerin tümü arındaki etkileşim net olrak gözükmez

**adı üstünde scatterplot matrisi** 

```python
sns.pairplot(idt)
```

saçılım grafikleri toz bukutu şeklindeyse veya yapısal bir formu yokssa bu o iki değişken arsında ilşki olmadığını gösterir

grafiklerinde yukarıdakş gibi kümleşen bir dağilim gözlemlendiğinde 

bir boyut daha eklenip (çaprazlama)bu kümleşmeleri anladıktan sonra yorum yapmka daha mantıklı olur

```python
sns.pairplot(idt, hue="species")
```

işaretleme şekilleri değiştirilebilir

```python
sns.pairplot(idt, hue="species", markers=["o","s", "D"])
```

## grafiğe doğru ekleme

```python
sns.pairplot(idt, hue="species", kind="reg")
```
