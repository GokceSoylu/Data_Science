```python
import pandas as pd
import seaborn as sns
data=sns.load_dataset("diamonds")
df=data.copy()
df.columns
```

```python
from pandas import CategoricalDtype
cut_ctg=["Fair","Good","Very Good","Premium","Ideal"]
df.cut=df.cut.astype(CategoricalDtype(ordered=True, categories=cut_ctg))
df.cut.head()
```

## catplot

* **catplot(x="değişken_adi", y="değişken adi, dat=data_frame)**

* seaborn fonksiyonu

* fonksiyonu barplot yada plot.barh gibi aitn grafiği oluşturanbir fonksiyon

hocam bu seaborn kütüphanesinin o olayı bu biliyorsun barplot fonksiyonunda da x y değiişkenlerini ve data yı alacağı yeri bildiiriyorduk :)


```python
sns.catplot(x="cut", y="price", data=df);
```

bir boyut da ekleyelim

```python
sns.barplot(x="cut",y="price", hue="color",data=df)
```

## doğrulama

burada cut ve color değerlerinin bir nevi ortalaması verirlir. çubukalr da standard sappmayı yani ortalmadan uzaklaşmayı gösteriri

o zaman bizde doğrulama için cut ve color a göre bir gruplama yaparız daha sonra bu bu gruplarddakş price değişkeninnin ortalamsını alır ve

grafikle örtüşüyor mu diye kontrol ederiz

```python
df.groupby(["cut","color"])["price"].mean()
```
