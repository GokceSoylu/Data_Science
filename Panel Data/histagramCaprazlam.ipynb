```python
import pandas as pd
import numpy as np
import seaborn as sns
data=sns.load_dataset("diamonds")
df=data.copy()
df.info()
```

```python
sns.kdeplot(df.price, fill=True);
```

burada mavi alanı oluşturan nedir?

ana hadef değişken burada price. 

```python
(sns
.FacetGrid(df,
    hue="cut",
    height=5,
    xlim=(0,10000))
.map(sns.kdeplot, "price", fill=True)
.add_legend()
);
```

```python
sns.catplot(x="cut", y="price",hue="color", data=df) #sözde boyut arttırdık üç değişkenli yaotık ama işe yaramadı okunaklı değil
#bu durumlarda gösterim şekli içn sütun değil nokta daha nma ntıklı olur
```

```python
sns.catplot(x="cut", y="price", hue="color", kind="point", data=df)
```
