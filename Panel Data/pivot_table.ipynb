```python
import pandas as pd
import numpy as np
import seaborn as sns
titanic=sns.load_dataset("titanic")
titanic.head(10)
```

## pivot suz

```python
titanic.groupby("sex")["survived"].mean()
```

```python
titanic.groupby("sex")[["survived"]].mean()#bir parantez daha atmak fancy kullanmak gibi oluyor.  
```

```python
titanic.groupby(["sex","class"])[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].mean().unstack()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate(["mean",np.sum])
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate(["mean",np.sum]).unstack()
```

## pivot

```python
titanic.pivot_table("survived",index="sex", columns="class")
```

```python
titanic.pivot_table("survived",index=["sex","who"], columns="class") #boyle çok boyutlu da yapabiliyorız
```
