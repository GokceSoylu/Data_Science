# Data frame birleştirme ilşlemleri 

## concat([data_farame1, data_frame2]) fonksiyounun kulllanırız

üç argüman daha alabilir. 

>ignore_index=bool index farklarını göz ardı eder ve süzgünce numaralandırır.

>join="inner"  kesişimleri alır

>join_axes=[data_frame.columns] burada da farklı değişken türü olmasına karşın belirli değişkene göre birleştirme yapma


```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame(np.arange(1,21,1).reshape(5,4))
df1.columns="a","b","c","d"
df1
```

```python
df2=df1+20
df2
```

```python
pd.concat([df1,df2])
#görüldüğü üzere yazılma sırasına göre ekler :))
```

```python
pd.concat([df1,df2],ignore_index=True)
#index farkı olmasın düzenli olsun istiyorsan ignore_index=True deriz.bu defoult False dır
#denedik ignore_column diye birşey yok :))
#todo df1.index="x","y","z","t","v" 
#bu şekilde index isimlerini değiştirirsen ve sonra birleştitirsen index ismi gözükür sorun olmaz
#ancak ignore_index=True dersen baştan kendisi klak indisleme yapar :))
```

```python
df1.index="x","y","z","t","v"
df1
```

```python
pd.concat([df1,df2],ignore_index=True) #!adding by column 
```

**peki column değişkenleri farklı olursa.**

işte o zaman sıkıntı yaşanır. çünkü column değeri sabit kalıp index artacak şeklide ekleme yapılır.

hepsini yazar ve diğerinde olmyan değşkene nan yazar. bunu yine ekliyiceğimiz argümanlarla düzenleyebiliriz.

reindex bunu data frame isimlerini yazdığımız parantezlerin içerisine yazarız. 

df1._reindex=(columns=df2.columns)_ deriz. burada birleştirme df1 nin column değerlerine göre yapılır.

**?peki bunu index için yapamaz mıyız** eveet yapabiliriz. _df1.reindex(index=df2.index)_ deriz df2 nin index değerlerine göre birleştirme yapılır 

```python
df1.columns="a","b","c","e"
df1
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1, df2.reindex(columns = df1.columns)])
```

```python
pd.concat([df1, df2.reindex(index = df1.index)])
```
