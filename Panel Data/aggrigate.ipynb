```python
import pandas as pd
import numpy as np
```

```python
a=pd.DataFrame({"a":[100,200,300,400],
                "b":[300,234,23,56],
                "c":[45,34,678,34],
                "gruplar":["var1","var2","var1","var2"]},columns=["gruplar", "a","b","c"])
a
```

```python
a.groupby("gruplar")
```

```python
a.groupby("gruplar").describe()
```

```python
a.groupby("gruplar").aggregate([np.median, np.sum, "mean"])
```
