### **barplot catplot**
aşşağıda grafikleri barplot ve catplot fonksiyonlari ile göreslleştirdim. yoğunluğu görmek ve özelliklede ikiden fazla 

değişkenin bulunduğu grafikler için catplot daha iyi. Bu drumlarda sütun grafiği(braplot) yetersiz kalıyor.

* ! barplot ve catplot en az bir tane numeric değişkne olmak zorunda !

```python
sns.catplot(x="sex", y="tip", data=tp)
```

```python
sns.barplot(x="sex", y="tip", data=tp)
```

```python
sns.catplot(x="time", y="tip", data=tp) 
```

```python
sns.catplot(x="time", y="tip", hue="sex",data=tp)
```

```python
sns.catplot(x="smoker",y="tip", hue="size", data=tp, kind="point")
```

```python
sns.barplot(x="smoker",y="tip", hue="size", data=tp)
```

```python
sns.catplot(x="size",y="tip", hue="sex", data=tp)
```

## **displot hisplot**

plot fonksiyonlarının tamamı için (x= ,y= ,hue= ,data )değişkenlerini kullanabilrisn. Farkalar

* catplot ve barplot fonksiyonları en az bir tane numeric değişkene ,htiyaç duyar

* kdeplot fonksiyonu hem x hem y eğişkeninin numeric olmasını şster!

```python
sns.histplot(tp.sex) #görüldüğü üzere sadece categorical değişken ğzerinden de grafik yapaniliyor
```

```python
sns.histplot(x="sex", y="tip", data=tp)
```

```python
sns.histplot(x="sex", y="tip", data=tp, kde=True)
```

```python
sns.histplot(x="time", y="tip", hue="sex",data=tp)
```

```python
sns.displot(x="time", y="tip", hue="sex",data=tp)
```

```python
sns.kdeplot(x="tip", y="size", hue="sex",data=tp, fill=True)
```
